```python
def normalize(undirected):
    sum_df = undirected.sum()
    sp_list = sum_df[sum_df > 0].index.tolist()
    undirected = undirected.loc[sp_list, sp_list]
    col_sum = np.sqrt(undirected.sum())
    undirected = undirected.div(col_sum, axis=1)
    undirected = undirected.div(col_sum, axis=0)
    undirected = util.log_rescale(undirected)
    return undirected
```

```python
level = 'g'
indir = '../data/ARG'
outdir = '../result/S3_ARG/network_feature'

if not os.path.exists(outdir):
    os.makedirs(outdir)

with open('../result/S1_plasmid_net/tree/log_rescale/gksb/renamed_GCN_tree.newick') as fp:
    newick_tree = fp.read()

data_dict = {}
for file in os.listdir(indir):
    if not file.endswith('_g.tsv'):
        continue
    infile = os.path.join(indir, file)
    dtype = file.split('_plasmid')[0].split('arg_')[-1]
    data = pd.read_csv(infile, sep='\t', header=0, index_col=0)
    data = normalize(data)
    data_dict[dtype] = dc(data)
    print(data.shape)
```

```python
import copy
from collections import Counter
common_df = pd.DataFrame()
occurrence_df = pd.DataFrame()
for k in report_dict.keys():
    df = copy.deepcopy(report_dict[k]['norm_df'])
    common_df, df = common_df.align(df, fill_value=0)
    common_df += df
    occurrence_df, df = occurrence_df.align(df, fill_value=0)
    occurrence_df += (df > 0).astype(int)

values = np.triu(occurrence_df.values, k=1)
counter = Counter(values.flatten())
edge_num = len(values.flatten()) - counter[0]
edge_freq = {}
for k, v in counter.items():
    v = v/edge_num
    edge_freq[k] = v
del edge_freq[0]

types = list(edge_freq.keys())
percentages = list(edge_freq.values())
```

```python
for k in report_dict.keys():
    df = copy.deepcopy(report_dict[k]['norm_df'])
    sp_list = list(df.columns)
    edge_n = 0
    common_n = 0
    for i in range(len(sp_list)):
        for j in range(i+1, len(sp_list)):
            if df.iloc[i, j] > 0:
                edge_n += 1
                if occurrence_df.loc[sp_list[i], sp_list[j]] > 1:
                    common_n += 1
    print(f'{k}: {common_n} {edge_n} {common_n/edge_n}')
```

```python
pair_dict = {}
for k in report_dict.keys():
    df = copy.deepcopy(report_dict[k]['norm_df'])
    df[df > 0] = 1
    pair_df = pd.DataFrame(index=occurrence_df.columns, columns=occurrence_df.columns)
    pair_df.fillna(0, inplace=True)
    pair_df, df = pair_df.align(df, fill_value=0)
    pair_df += df
    pair_dict[k] = dc(pair_df.loc[occurrence_df.columns, occurrence_df.columns])
```

```python
common_rate_df = pd.DataFrame(index=report_dict.keys(), columns=report_dict.keys())
klist = list(report_dict.keys())
for i in range(len(klist)):
    k1 = klist[i]
    for j in range(i+1, len(klist)):
        k2 = klist[j]
        union = pair_dict[k1] + pair_dict[k2]
        union[union>0] = 1
        intersection = pair_dict[k1] * pair_dict[k2]
        common_rate = intersection.sum().sum()/union.sum().sum()
        common_rate_df.loc[k1, k2] = common_rate
        # print(k1, k2, common_rate)

common_rate_df.fillna(0, inplace=True)
common_rate_df += common_rate_df.T
np.fill_diagonal(common_rate_df.values, np.nan)
sum = common_rate_df.sum(axis=1)
sorted_index = sum.sort_values(ascending=False).index
mask = np.zeros_like(common_rate_df)
mask[np.triu_indices_from(common_rate_df)] = True
sns.heatmap(common_rate_df.loc[sorted_index, sorted_index], cmap='coolwarm', square=True)
#sns.clustermap(common_rate_df, cmap='coolwarm', square=True, metric ='jaccard')
#plt.tight_layout()
plt.savefig(os.path.join(outdir, 'common_rate.svg'), format='svg', bbox_inches='tight')
```
